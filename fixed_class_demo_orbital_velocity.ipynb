Writing formula to calculate orb vel

```python
def calculate_orbital_velocity(mass, radius):
  return(np.sqrt(ac.G * mass / radius))
```

Testing formula with earth mass and earth radius in km/s

```python
mass = 1 * u.earthMass
radius = 1 *u.earthRad
orb_vel = (calculate_orbital_velocity(mass, radius))
print(orb_vel.to(u.km / u.s))
```

Calculating 1 or 28 orb vel by typying in information

```python
mass = 3e7 * u.solMass
radius = 5.68 * 1000 * u.parsec
orb_vel = calculate_orbital_velocity(mass, radius)
print(orb_vel.to(u.km / u.s))
```

writing a code for all 28 orb vel to be solved

```python
vel_bh = np.zeros(np.shape(dis)) * u.km / u.s

for i, radius in enumerate (dis):
  rad = radius * 1000 * u.parsec
  orb_vel = calculate_orbital_velocity(mass, rad)
  vel_bh[i] = orb_vel
  print("The orbital velocity at ", radius, " kpc= ", orb_vel.to(u.km / u.s))
```

```python
def calculate_disk_velocity(disk_density, radius):
  return(np.sqrt( ac.G * disk_density * np.pi * radius ))
```

```python
def calculate_halo_velocity(halo_density, radius):
  return(np.sqrt( ac.G * halo_density * 4/3 * np.pi * radius ** 2 ))
```

```python
halo_mass = 4.76e10 * u.solMass
halo_radius = 35e3 * u.parsec
test_radius = 20 * 1000 * u.parsec
halo_density = (halo_mass)/ (np.pi * 4/3 * (halo_radius) ** 3)
halo_velocity = calculate_halo_velocity(halo_density, test_radius)
print (halo_velocity.to(u.km / u.s))
```

```python
vel_halo = np.zeros(np.shape(dis)) * u.km / u.s

for i, radius in enumerate (dis):
  rad = radius * 1000 * u.parsec
  orb_vel = calculate_halo_velocity(halo_density, rad)
  vel_halo[i] = orb_vel
  print("The orbital velocity at ", radius, " kpc= ", orb_vel.to(u.km / u.s))
```

```python
def calculate_total_velocity(radius, disk_density, halo_density, bulge_mass, blackhole_mass):
  disk_mass = disk_density * (np.pi * radius ** 2)
  halo_mass = halo_density * ( 4/3 * np.pi * radius ** 3)
  return np.sqrt( ac.G * (disk_mass + halo_mass + mass_bulge + blackhole_mass) / radius)
```

```python
vel_total = np.zeros(np.shape(dis)) * u.km / u.s
bh_mass = 3e7 * u.solMass
bulge_mass = 3e10 * u.solMass
for i, radius in enumerate (dis):
  rad = radius * 1000 * u.parsec
  orb_vel = calculate_total_velocity(rad, disk_density, halo_density, bulge_mass, bh_mass )
  vel_total[i] = orb_vel
  print("The orbital velocity at ", radius, " kpc= ", orb_vel.to(u.km / u.s))
```
